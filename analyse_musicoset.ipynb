# analyse temporelle (distribution des sorties des albums)

```python
import matplotlib.pyplot as plt

df_albums = df_albums.rename(columns={'name': 'album_name'})

df_combined = pd.merge(df_releases, df_artists, how='inner', on='artist_id')
df_combined = pd.merge(df_combined, df_albums, how='inner', on='album_id')

df_artist_albums = df_combined[df_combined['name'] == "David Guetta"]

plt.figure(figsize=(10, 6))
plt.scatter(df_artist_albums['release_date'], df_artist_albums['album_name'], color='skyblue')
plt.title('Diagramme de dispersion des sorties d\'albums de Eminem')
plt.xlabel('Date de sortie')
plt.ylabel('Album')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Lecture du fichier CSV des artistes
df_artists = pd.read_csv("musicoset/artists.csv", sep="\t")

# Assurez-vous que les colonnes popularity et followers sont de type numérique
df_artists['popularity'] = pd.to_numeric(df_artists['popularity'], errors='coerce')
df_artists['followers'] = pd.to_numeric(df_artists['followers'], errors='coerce')

# Créer un nuage de points interactif avec Plotly
fig = px.scatter(df_artists, x='popularity', y='followers', color='followers',
                 title='Corrélation entre Popularité et Followers',
                 labels={'popularity': 'Popularité de l\'artiste', 'followers': 'Nombre de followers'})

# Ajouter une échelle de couleur basée sur le nombre de followers
fig.update_traces(marker=dict(size=8, opacity=0.5))
fig.show()
```

```python
import seaborn as sns

# Créer un nuage de points avec ligne de régression
plt.figure(figsize=(10, 6))
sns.regplot(x='popularity', y='followers', data=df_artists, scatter_kws={'color': 'skyblue', 'alpha':0.5})
plt.title('Corrélation entre Popularité et Followers avec Régression linéaire')
plt.xlabel('Popularité de l\'artiste')
plt.ylabel('Nombre de followers')
plt.grid(True)
plt.show()
```

# artiste avec le plus grand nombre de feat

```python
import re

collaboration_counts = df_songs[df_songs['song_type'] == 'Collaboration']['artists']
collaboration_counts

dic_count = {}

for row in collaboration_counts:
    row = row.replace("{", "")
    row = row.replace("}", "")
    split_data = row.split(",")
    for split in split_data:
        #regex pour récupérer uniquement la partie qui m'interesse
        match = re.search(r":\s*'([0-9A-Za-z\s.-]+)'", split)
        if match:
            artist_name = match.group(1)
            if artist_name not in dic_count:
                dic_count[artist_name] = 0
            else:
                dic_count[artist_name] += 1

sorted_dic_count = dict(sorted(dic_count.items(), key=lambda x: x[1], reverse=True))
sorted_dic_count
```
